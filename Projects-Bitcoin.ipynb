## 3.1 Assets Diversification
Asset diversification is a fundamental principle of modern investment strategy, notable for managing risk and optimizing returns. Rooted in fundamental financial theory, diversification’s primary goal is to mitigate the risks associated with overconcentration in any single asset or market. This strategy involves spreading investments across multiple asset classes such as equities, bonds, real estate, and commodities, leveraging the lack of perfect correlation among various market sectors to reduce the overall volatility of the portfolio and enhance the potential for stable returns over time. Extensive research over the decades has continuously reinforced the importance of diversification, validating its role as a critical component of prudent investment management.

The concept of diversification is firmly embedded in Harry Markowitz's Modern Portfolio Theory (MPT), introduced in the 1950s. MPT demonstrated how diversification could substantially reduce the volatility of a portfolio while maintaining or enhancing returns. This theory posits that an efficiently diversified portfolio maximizes returns for any given level of risk. This foundational principle has been empirically validated in numerous studies, indicating that a well-diversified portfolio typically outperforms a concentrated one over a long period. This success is primarily due to diversification allowing investors to mitigate idiosyncratic risks—the risks inherent to particular investments—thereby protecting the portfolio against significant losses that might arise from any single investment underperforming (Laopodis, 2020).

Effective diversification involves more than just mixing different stocks; it extends to including a variety of asset classes such as bonds, commodities, real estate, and potentially alternative investments like hedge funds and private equity. Each asset class reacts differently to economic events; for example, bonds might perform well during times when stocks are declining, and commodities such as gold are often seen as safe havens during market turmoil. This strategy not only buffers the portfolio against sector-specific downturns but also positions it to capitalize on the growth of multiple sectors across different economic cycles. Detailed statistical analysis has shown that strategic asset allocation, determining the proportionate investment in each asset class, can explain the majority of a portfolio’s return variability, highlighting the critical role of diversification in achieving optimal investment outcomes (Choi, 2011).

Śmietana (2014) discussed that incorporating real estate into a diversified portfolio can significantly enhance its mean-variance efficiency. Real estate possesses unique characteristics such as high tangible asset value, relatively stable income generation through rents, and potential tax advantages. Additionally, real estate often exhibits low correlation with stocks and bonds, thereby providing substantial diversification benefits. Studies have demonstrated that mixed-asset portfolios that include real estate assets not only yield higher returns but also maintain lower volatility compared to portfolios comprising solely financial assets. This supports the value of broadening the investment horizon to include real estate as a fundamental element of a diversified investment strategy.

The fluctuations of financial assets, a critical measure for investors and analysts, has been extensively explored in traditional markets. However, the advent of digital currencies, particularly Bitcoin, has expanded the discussion to include these new assets, which are known for their distinctive price fluctuations.

## 4 Theoretical Framework:

The **Efficient Market Hypothesis (EMH)**, initially proposed by **Fama (1970)**, posits that asset prices fully reflect all available information at any given time. According to EMH, securities are always traded at their fair value on exchanges, making it impossible for investors to consistently achieve returns that exceed average market returns on a risk-adjusted basis, except through random chance. The hypothesis is typically expressed in its weak, semi-strong, and strong forms, each assuming a different level of market efficiency based on the scope of information reflected in market prices:

- **Weak-form efficiency** asserts that all past market prices and data are fully reflected in current asset prices. Thus, technical analysis of past price movements would not be expected to yield predictive power regarding future price movements.
- **Semi-strong form efficiency** contends that all publicly available information is reflected in asset prices, not just past prices. Under this form, neither fundamental analysis (which examines financial statements, news reports, economic forecasts, etc.) nor technical analysis would be expected to produce superior returns.
- **Strong-form efficiency** argues that all information, both public and private (inside information), is fully accounted for in prices. Therefore, no one can achieve superior returns.

In the context of **Bitcoin** and other cryptocurrencies, the application of EMH presents new challenges. The speculative nature of Bitcoin, influenced by its decentralized structure and the influence of non-traditional factors such as technological changes, community sentiment, and geopolitical events, often leads to significant price fluctuations that appear to defy traditional market logic.

## 6 Methdology



This study employs a quantitative approach to analyze and compare the volatility of Bitcoin and the S&P 500, utilizing the Generalized Autoregressive Conditional Heteroskedasticity (GARCH) model. Specifically, the GARCH(1,1) model is chosen for its effectiveness in capturing the "clustering" of volatility, a common phenomenon in financial time series data where periods of high volatility are followed by similar high volatility periods, and similarly for low volatility.

The decision to use the GARCH(1,1) model is supported by numerous studies demonstrating its robustness in financial applications. Engle (2001), who introduced the concept of autoregressive conditional heteroskedasticity, highlights the GARCH model's capacity to provide detailed insights into financial market volatility dynamics. Additionally, Hansen and Lunde (2005) conducted comparisons among various GARCH models and found that the GARCH(1,1) model frequently performs comparably or better than more complex models. This makes it a preferred choice for financial data analysis due to its simplicity and effectiveness.

The model parameters estimated include omega (ω), which represents the baseline volatility level; alpha (α), measuring the response of volatility to shocks from the previous day; and beta (β), indicating the persistence of volatility from one day to the next.

The analysis involved first cleaning and preprocessing the data, converting price data into returns, and then applying the GARCH model to these return series. The estimated parameters from the GARCH model provide insights into the volatility characteristics of each asset, which are crucial for understanding and managing investment risks in these markets.


This methodology aligns with current financial econometric practices and addresses the need for understanding asset volatility for better risk management and investment decision-making, as demonstrated in the seminal works of Bollerslev (1986) and Engle and Bollerslev (1986).

### Regression Form of GARCH(1,1)

The GARCH(1,1) model can be represented as follows in a regression-like format:

1. **Return Equation**:
   $$
   r_t = \mu + a_t
   $$
   Here, $r_t$ represents the return of the asset at time $t$, $\mu$ is the constant mean of the returns, and $a_t$ is the residual term at time $t$.

2. **Volatility (Conditional Variance) Equation**:
   $$
   \sigma_t^2 = \omega + \alpha a_{t-1}^2 + \beta \sigma_{t-1}^2
   $$
   In this equation:
   - $\sigma_t^2$ denotes the conditional variance (or volatility) of the returns at time $t$.
   - $\omega$ (omega) represents the long-run average variance.
   - $\alpha$ (alpha) captures the responsiveness of the current variance to the squares of the lagged (previous period’s) residuals, reflecting the impact of the last period's shocks on today's volatility.
   - $\beta$ (beta) measures the persistence of past variances, indicating how much of yesterday's variance is remembered today.

A fundamental assumption underlying the application of GARCH models is stationarity. This means that the statistical properties of the series—mean, variance, and autocorrelation—must not depend on the time at which the series is observed. However, financial time series, including Bitcoin returns, often exhibit properties that violate stationarity, such as trends or heteroskedasticity.


The stationarity problem in GARCH models is critical because the lack of it can lead to unreliable and biased estimates, making the model's forecasts less accurate and potentially misleading. For a GARCH model to produce valid and consistent results, the conditional mean and variance of the series must be strictly stationary. However, Bitcoin returns can show non-stationary behaviors like unit roots, structural breaks, or seasonal effects, which must be addressed before fitting a GARCH model. Addressing the issue of stationarity involves testing the Bitcoin return series for stationarity using techniques like the Augmented Dickey-Fuller (ADF) test. Next, we will perform an ADF test to see whether it is stationary.


```python
import pandas as pd
from arch import arch_model
from statsmodels.tsa.stattools import adfuller, acf
from statsmodels.stats.diagnostic import acorr_ljungbox

# Load your data
bitcoin_data_path = "E:\\611 Final Project\\bitcoin_data.csv"
bitcoin_data['Returns'] = bitcoin_data['Close'].pct_change().dropna()

# Test for stationarity on returns
adf_result = adfuller(bitcoin_data['Returns'].dropna())
print("ADF Statistic: %f" % adf_result[0])
print("p-value: %f" % adf_result[1])
```

From the ADF test we can see series is non-stationary. Onne common method to achieve stationarity is differencing the series. By taking the first or second differences of Bitcoin returns, the non-stationary trend components may be eliminated, leaving a stationary series suitable for GARCH modeling.
Also we can log difference the series to achieve stationary. 

```python
##now we will difference the series
bitcoin_data['Differenced_Returns'] = bitcoin_data['Returns'].diff().dropna()

# Test for stationarity on the differenced returns
adf_result_diff = adfuller(bitcoin_data['Differenced_Returns'].dropna())
print("ADF Statistic for Differenced Returns: %f" % adf_result_diff[0])
print("p-value for Differenced Returns: %f" % adf_result_diff[1])

# Decision based on p-value
if adf_result_diff[1] > 0.05:
    print("Differenced series is still not stationary. Further investigation may be needed.")
else:
    print("Differenced series is now stationary.")
```

To further confirm the result of the series stationary, a Time Series Plot with Rolling Statistics and an ACF plot is helpful.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Assuming your DataFrame is loaded and the 'Returns' column exists
# Load your data if not already loaded
# bitcoin_data = pd.read_csv('path_to_your_bitcoin_data.csv')
# bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])

# Calculate Differenced Returns if not already calculated
bitcoin_data['Differenced_Returns'] = bitcoin_data['Returns'].diff().dropna()

# Plotting the differenced returns and their rolling statistics
fig, ax = plt.subplots(figsize=(12, 6))
bitcoin_data['Differenced_Returns'].plot(ax=ax, label='Differenced Returns', color='darkblue')  # Change color here
rolling_mean = bitcoin_data['Differenced_Returns'].rolling(window=12).mean()
rolling_std = bitcoin_data['Differenced_Returns'].rolling(window=12).std()

rolling_mean.plot(ax=ax, color='green', label='Rolling Mean')  # Change color here
rolling_std.plot(ax=ax, color='magenta', label='Rolling Std')  # Change color here
ax.set_title('Differenced Returns with Rolling Mean & Standard Deviation')
ax.set_xlabel('Date')
ax.set_ylabel('Differenced Returns')
ax.legend()
plt.show()

# Plotting the Autocorrelation Function (ACF)
fig, ax = plt.subplots(figsize=(12, 6))
plot_acf(bitcoin_data['Differenced_Returns'].dropna(), ax=ax, lags=40, color='darkred')  # Change color here
ax.set_title('Autocorrelation Function of Differenced Returns')
plt.show()
```

In the first plot, the differenced returns are visualized along with their rolling mean and standard deviation, calculated over a 12-period window. The rolling mean (in green) remains consistently around zero, while the rolling standard deviation (in magenta) exhibits some fluctuations but generally remains stable, indicating that the differenced returns possess a stable variance over time. Such characteristics suggest that the time series is stationary, with no evident long-term trends or cyclical patterns in the mean and variance.

The second plot, the Autocorrelation Function (ACF), shows the autocorrelations of the differenced returns at various lags. The autocorrelations for almost all lags lie within the 95% confidence interval and quickly taper off towards zero, indicating a lack of significant autocorrelation. This rapid decline and clustering around zero further support the stationarity of the series, as it suggests that past values have minimal influence on future values. Together, these plots affirm that the differenced returns do not exhibit time-dependent structures, making them suitable for the GARCH (1,1) model.

The ADF statistic of -9.197357 is much lower than the critical values at all common significance levels (1%, 5%, and 10%). This suggests a very strong rejection of the null hypothesis. The test statistic is more negative than any of the critical values, this strongly supports the conclusion that the series is stationary.

```python
import pandas as pd

# Load the S&P 500 dataset
sp500_data_path = "E:\\611 Final Project\\S&P_500_2.csv"
sp500_data = pd.read_csv(sp500_data_path)

# Convert 'Date' to datetime format
sp500_data['Date'] = pd.to_datetime(sp500_data['Date'])

# Replace commas and convert 'Price', 'Open', 'High', and 'Low' to numeric values
price_columns = ['Price', 'Open', 'High', 'Low']
for col in price_columns:
    sp500_data[col] = pd.to_numeric(sp500_data[col].str.replace(',', ''), errors='coerce')

# Calculate daily returns from the 'Price' column
sp500_data['Returns'] = sp500_data['Price'].pct_change()

# Drop the first row since it will have a NaN value for returns and any other possible NaN values
sp500_data = sp500_data.dropna()

# Perform first differencing on the 'Returns' column
sp500_data['Differenced_Returns'] = sp500_data['Returns'].diff().dropna()

# Check the first few entries of the differenced series
print(sp500_data[['Date', 'Returns', 'Differenced_Returns']].head())
```

Checking for stationary again: 

```python
# Perform the ADF test again on the differenced series
adf_test_differenced = adfuller(sp500_data['Differenced_Returns'].dropna())
print("ADF Statistic on Differenced Series: %f" % adf_test_differenced[0])
print("p-value: %f" % adf_test_differenced[1])
print("Critical Values:")
for key, value in adf_test_differenced[4].items():
    print('\t%s: %.3f' % (key, value))
```

The differenced series has an ADF statistic of -8.92343, which is significantly lower than all the threshold values for the 1%, 5%, and 10% critical values. This indicates a strong rejection of the null hypothesis. Given that the ADF statistic is well below the critical values, it further supports the stationarity of the differenced returns series.

To further confirm the result of the series stationary, a Time Series Plot with Rolling Statistics and an ACF plot is helpful.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Assuming your DataFrame is loaded and the 'Returns' column exists
# Calculate Differenced Returns if not already calculated
sp500_data['Differenced_Returns'] = sp500_data['Returns'].diff().dropna()

# Plotting the differenced returns and their rolling statistics
fig, ax = plt.subplots(figsize=(12, 6))
sp500_data['Differenced_Returns'].plot(ax=ax, label='Differenced Returns')
rolling_mean = sp500_data['Differenced_Returns'].rolling(window=12).mean()
rolling_std = sp500_data['Differenced_Returns'].rolling(window=12).std()

rolling_mean.plot(ax=ax, color='red', label='Rolling Mean')
rolling_std.plot(ax=ax, color='black', label='Rolling Std')
ax.set_title('Differenced Returns with Rolling Mean & Standard Deviation')
ax.set_xlabel('Date')
ax.set_ylabel('Differenced Returns')
ax.legend()
plt.show()
# Plotting the Autocorrelation Function (ACF)
fig, ax = plt.subplots(figsize=(12, 6))
plot_acf(sp500_data['Differenced_Returns'].dropna(), ax=ax, lags=40)
ax.set_title('Autocorrelation Function of Differenced Returns')
plt.show()
```

From the above plot, the rolling standard deviation appears fairly constant over time, though it does show some variability corresponding to the spikes in the differenced returns. The constancy in the rolling standard deviation overall supports the idea that the series has a constant variance, another criterion for stationarity.

Also, the ACF plot for the differenced returns shows that autocorrelations at most lags are insignificant (within the 95% confidence interval), which is indicative of a white noise series. This further supports the stationarity of the series, as the absence of autocorrelation is a strong indicator of a stationary process.

```python
# Load the S&P 500 dataset
sp500_data_path = "E:\\611 Final Project\\S&P_500_2.csv"
sp500_data = pd.read_csv(sp500_data_path)
# Convert 'Date' to datetime format
sp500_data['Date'] = pd.to_datetime(sp500_data['Date'])

# Replace commas and convert 'Price', 'Open', 'High', and 'Low' to numeric values
price_columns = ['Price', 'Open', 'High', 'Low']
for col in price_columns:
    sp500_data[col] = pd.to_numeric(sp500_data[col].str.replace(',', ''), errors='coerce')

# Calculate daily returns from the 'Price' column
sp500_data['Returns'] = sp500_data['Price'].pct_change() * 100

# Drop the first row since it will have a NaN value for returns and any other possible NaN values
sp500_data = sp500_data.dropna()
import pandas as pd
from arch import arch_model


# Assuming you have a pandas DataFrame 'sp500_data' with a column 'Returns' for the return series
returns = sp500_data['Returns']

# Define the GARCH(1,1) model
garch11_model = arch_model(returns, vol='Garch', p=1, q=1)

# Fit the model
garch11_results = garch11_model.fit(update_freq=10)  # update_freq=10 for less verbose output

# Print the summary of the model fit
print(garch11_results.summary())

# Optional: Convert the summary to a DataFrame for better visual representation in Jupyter
# Creating a DataFrame from the summary
results_df = pd.DataFrame({'Parameter': garch11_results.params.index,
                           'Coefficient': garch11_results.params.values,
                           'Std. Error': garch11_results.std_err.values,
                           't-values': garch11_results.tvalues.values,
                           'P>|t|': garch11_results.pvalues.values,
                           'Conf. Interval Lower': garch11_results.conf_int().iloc[:, 0],
                           'Conf. Interval Upper': garch11_results.conf_int().iloc[:, 1]})

# Apply styling to the DataFrame for better readability
styled_results = results_df.style.format({
    'Coefficient': "{:.4f}",
    'Std. Error': "{:.4f}",
    't-values': "{:.2f}",
    'P>|t|': "{:.4f}",
    'Conf. Interval Lower': "{:.4f}",
    'Conf. Interval Upper': "{:.4f}"
}).background_gradient(cmap='coolwarm', subset=['Coefficient', 't-values'])

# Display the styled DataFrame
styled_results

```

- **Omega (Baseline Volatility):** The estimated baseline volatility, when past shocks and volatility are not considered, is 0.0382. This parameter is statistically significant with a p-value of 0.00571, implying a non-zero baseline level of volatility in the S&P 500 returns.

- **Alpha:** The estimated value is 0.3303, which is statistically significant with a p-value of 0.01491. This value indicates that past shocks have a significant effect on current volatility, suggesting that volatility reactions to new information are relatively pronounced.

- **Beta:** The estimated value is 0.5796 and is highly significant with a p-value virtually at 0. This high and significant beta indicates strong persistence of volatility over time, meaning that the effects of past volatility are likely to persist into the future.

In GARCH models, the focus is on modeling the variance (volatility) of a time series rather than the time series itself, which is why the R squared value, commonly used in traditional regression analysis to measure the proportion of variance explained by the independent variables, is not typically a meaningful statistic in this context. Engle (1982) introduces the ARCH model in his seminal paper, which is the foundation for GARCH models. Engle's work primarily focuses on the conditional variance rather than the conditional mean, and he elaborates on how traditional metrics like R squared are not adequate for evaluating the goodness of fit for these models. The focus is rather on the ability of the model to capture the dynamics of volatility based on lagged error terms.

Also the result from the GARCH (1,1) critically engage with the Efficient Market Hypothesis (EMH). Specifically, the GARCH(1,1) parameters, omega (ω), alpha (α), and beta (β), furnish empirical insights that question the market's adherence to both the weak and semi-strong forms of EMH. Under the weak form of EMH, past price volatility should not predict future volatility; however, the significant beta (β) parameter in our model suggests a strong persistence of volatility over time. This persistence indicates that previous periods of high volatility are likely to be followed by similarly high volatility, contrary to what would be expected if the market were efficiently processing and incorporating all historical price information into current values.

Moreover, the alpha (α) parameter measures the response of current volatility to shocks from immediate past returns. A significant alpha implies that recent unexpected price changes—presumably reflecting new information—have a substantial impact on current market volatility. This finding is pivotal, as it challenges the semi-strong form of EMH, which posits that all publicly available information is promptly reflected in asset prices, thereby negating any predictive advantage from analyzing past price movements. The empirical evidence presented by the GARCH model suggests that Bitcoin's price adjustments to new information are neither instant nor complete, thereby allowing past shocks to disproportionately influence future volatility.

The theoretical implications of these results are significant. They suggest that Bitcoin does not fully exhibit the characteristics of an efficient market as defined by EMH. This inefficiency could stem from a variety of factors intrinsic to Bitcoin’s market structure, such as its speculative nature, regulatory uncertainties, and its evolving acceptance as a mainstream financial asset. Consequently, these results not only challenge the conventional understanding of market efficiency in the context of digital currencies but also underscore the necessity for investors to consider strategies that account for predictable patterns in volatility when making investment decisions.
